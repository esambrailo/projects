```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas

%matplotlib inline
```

---
#### 1) List of Country Files

```python
import os

folder = './country_data/' # folder path
country_files = [] # list to store files

for file in os.listdir(folder): #iterates through each item in the specified folder
    if os.path.isfile(os.path.join(folder, file)): #checks if it is a file
        country_files.append(file) #appends to list
country_files.sort() #sorts the files in alphabetical order
```

---
#### 2) Dictionary of DataFrames
 - Creates a dataframe for the baseline and anomoly temperatures for each file
 - Appends dataframe to corresponding dictionaries

```python
import countrytemps

country_anoms = {}
country_bases = {}
countries = []
for country in country_files:
    country_anoms.update({country.replace('.txt',''): countrytemps.country_anomoly(country)})
    country_bases.update({country.replace('.txt',''): countrytemps.country_baselines(country)})
    country = countries.append(country.replace('.txt',''))
```

```python
#test sample set
sample_anoms = {}
sample_bases = {}
sample = []
for country in country_files[0:15]:
    sample_anoms.update({country.replace('.txt',''): countrytemps.country_anomoly(country)})
    sample_bases.update({country.replace('.txt',''): countrytemps.country_baselines(country)})
    country = sample.append(country.replace('.txt',''))
```

---
#### 3) Start/End Dates
 - Creating a new frame of just the first and last date for each country. 
 - Plotting histogram of those dates. 

```python
daterange = pd.DataFrame(columns = ['Country', 'Min_Yr', 'Max_Yr', 'Nulls'])
for country, df in country_anoms.items():
    new_row = pd.Series({'Country': country, 'Min_Yr': df.Year.min(), 'Max_Yr': df.Year.max(), 'Nulls': df.Monthly_Anomaly.isna().sum()})
    daterange = pd.concat([daterange, new_row.to_frame().T], ignore_index=True).sort_values('Min_Yr')

fig, ax = plt.subplots(1, 2, gridspec_kw={'width_ratios': [8, 1]}, figsize=(20, 5))
ax[0].hist(daterange.Min_Yr, bins = 200)
ax[0].set_title("Year Country Datasets Begin") #adding title
ax[0].set_xlabel("First Year In Datasets") #adding x axis label
ax[0].set_ylabel("Number of Countries") #adding y axis label
ax[0].tick_params(axis='x', labelrotation = 0)

ax[1].hist(daterange.Max_Yr, bins = 10)
ax[1].set_title("Year Country Datasets End") #adding title
ax[1].set_xlabel("Last Year In Datasets") #adding x axis label
ax[1].set_ylabel("Number of Countries") #adding y axis label
ax[1].tick_params(axis='x', labelrotation = 0)
pass
```

---
#### 4) Active Monthly Dates
 - Realized in previous exercise that there are gaps in the data in between the first and last dates present.  So every date for a country needs to be assessed. 
 - Creating a new DataFrame that contains every year for every country where that country had monthly data available.
 - Plotting a scatter plot representation for each country
 - Plotting a histogram of countries present by year. 

```python
active_dates = pd.DataFrame()
for country, df in country_anoms.items():
    for year in list(df.Year.unique()):
        new_row = pd.Series({'Country': country, 'Year': year})
        active_dates = pd.concat([active_dates, new_row.to_frame().T], ignore_index=True)
```

```python
fig1, ax1 = plt.subplots(1, 1, figsize = (12,40))
ax1.scatter(active_dates.Year, active_dates.Country, s=1)
ax1.set_title("Monthly Temperature Data Available\n(by Year and Country)") #adding title
ax1.set_xlabel("Year") #adding x axis label
ax1.set_ylabel("Country") #adding y axis label
ax1.tick_params(axis='x', labelrotation = 90)
ax1.margins(y=.005)
```

#### 5) Defining Date-Range
 - From the above exercise it appears that the year 1900 is a good starting point for temperature data from countries and potentially 2016 as the end date.
 - Doing further review of the data at those two time periods. 

```python
#comparing the country list from the defined start and stop years for the daterange
countries_2016 = list(active_dates[active_dates.Year == 2016].Country)
countries_1900 = list(active_dates[active_dates.Year == 1900].Country)

#identifying the countries that were added within our specified daterange
new_countries = pd.Series() 
for country in countries_2016:
    if country not in countries_1900:
        new_countries = pd.concat([new_countries, pd.Series(country).T], ignore_index=True)

new_countries
```

#### 6) Geospatial Mapping
 - Reviewing the available world geopandas dataset. 

```python
#reading in one of geopandas available datasets and reviewing
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
```

---
#### 7) Cross-Referencing Countries
 - Creating a dictionary that holds the naming convention for countries between the two datasets.
 - Reviewing the countries that are not in both files. 

```python
#iterating through the two different datasets to create a cross reference dictionary of countries in both
missing_countries = []
country_pairs = {} #dict to hold reference. key is from geopandas, value is from temp data
for country in world.name:
    if country in countries_2016:
        country_pairs[country] = country
    elif country.replace(" ","-") in countries_2016:
        country_pairs[country] = country.replace(" ","-")
    else:
        missing_countries.append(country)
        country_pairs[country] = None
```

```python
#manual pairing of country names between the two datasets 
country_pairs.update({'Bosnia and Herz.': 'Bosnia-and-Herzegovina', 
                      'Central African Rep.': 'Central-African-Republic', 
                      'Czechia': 'Czech-Republic', 
                      'Dem. Rep. Congo': 'Congo-(Democratic-Republic-of-the)',
                      'Dominican Rep.': 'Dominican-Republic',
                      'Eq. Guinea': 'Equatorial-Guinea',
                      'Falkland Is.': 'Falkland-Islands-(Islas-Malvinas)',
                      'Fr. S. Antarctic Lands': 'French-Southern-and-Antarctic-Lands',
                      'Palestine': 'Palestina',
                      'Solomon Is.': 'Solomon-Islands',
                      'United States of America': 'United-States',
                      'W. Sahara': 'Western-Sahara',
                     'eSwatini': 'Swaziland',
                     'Myanmar': 'Burma'})
```

```python
len(country_pairs) #number of countries that exist in geopandas dataset
```

```python
#identifying which countries still exist in geopandas that we don't have in our temp dataset
missing_countries = []
for country in country_pairs.keys():
    if country_pairs[country] == None:
        missing_countries.append(country)
        
print("geocountries not in our dataframes: ", len(missing_countries))
missing_countries
```

```python
#identifying what countries we have in our temp dataset that geopandas does not have
xtra_countries = []
for country in countries_2016:
    if country not in country_pairs.values():
        xtra_countries.append(country)
        
print("Additional countries in our dataframes: ", len(xtra_countries))
```

---
#### 8) Map of Countries with Data
 - Creating a map showing the countries that we have temperature data for. 

```python
#adding columns to the geopandas dataframe for the cross-referenced country name from the temp dataset
#adding another boolean column to identify whether we have a match
world['Temp_Name'] = world['name'].map(country_pairs)
world['Temp_Present'] = ~world.Temp_Name.isnull()
world.Temp_Present.value_counts()
```

```python
#testing how to derive the x,y coordinates from the geometry data for annotation
#adding a new column to the dataframe to hold the derived coordinates 
cntry = 'Albania'
world['coords'] = world['geometry'].apply(lambda x: x.representative_point().coords[:])
world['coords'] = [coords[0] for coords in world['coords']]
world[world.name == cntry]
```

```python
#plotting the map of countries with colors indicating what we do and dont have temperature data for 
fig3, ax3 = plt.subplots(1, 1, figsize = (15,15))
world.plot(column = 'Temp_Present', 
           ax=ax3, 
           cmap='prism', 
          categorical = True,
          legend = True)
ax3.set_title("Countries with Temperature Data Available") #adding title
ax3.set_xlabel("Longitude") #adding x axis label
ax3.set_ylabel("Latitude") #adding y axis label
ax3.xaxis.set_major_formatter('{x:1.0f} Â°')
ax3.yaxis.set_major_formatter('{x:1.0f} Â°')
ax3.margins(x=0.01, y=0.01)
ax3.patch.set_facecolor('blue')
ax3.patch.set_alpha(.05)

ax3.axhline(y = 0, color='grey', linestyle = ':') #line indicating the starting point of our data selection
ax3.text(-150, 1, 'Equator')
#providing annotation for each country that we do not have temperature data for
world[~world.Temp_Present].apply(lambda x: ax3.annotate(text=x['name'],
                                                   xy=x.geometry.centroid.coords[0],
                                                   xytext = (x.geometry.centroid.coords[0][0]+5,x.geometry.centroid.coords[0][1]+12),
                                                   arrowprops = {'facecolor': 'black', 'width':.001},
                                                   ha='center'), axis=1);
```

#### 9) Heat Map
 - Geospatially mapping the temperature variance between Jan. 1905 and Dec. 2011 for countries. 
 - Using the ten-year anomoly for metric. 

```python
#capture 'Ten_year_Anomaly' data for each country at Jan. 1905 & Dec. 2011
#quantify the null values that are retrieved. 
ten_yr_diff = pd.DataFrame()
for country, df in country_anoms.items():
    new_row = pd.Series({'Temp_Name': country, 
                         'Jan_1905': float(df.Ten_year_Anomaly[(df.Year == 1905) & (df.Month == 1)].sum()), 
                         'May_2011': float(df.Ten_year_Anomaly[(df.Year == 2011) & (df.Month == 5)].sum())})
    ten_yr_diff = pd.concat([ten_yr_diff, new_row.to_frame().T], ignore_index=True)

no_start = list(ten_yr_diff.Temp_Name[ten_yr_diff.Jan_1905 == 0])
no_end = list(ten_yr_diff.Temp_Name[ten_yr_diff.May_2011 == 0])
print('Countries without temp data in Jan 1905:\n', no_start)
print('Countries without temp data in May 2011:\n', no_end)
```

```python
#create calculated field for comparison start & end values

ten_yr_diff['Difference'] = np.where(ten_yr_diff.Jan_1905 != 0,
                                     (ten_yr_diff.May_2011 - ten_yr_diff.Jan_1905),
                                     'NaN')
```

```python
#append these values to the geopandas df
world = pd.merge(world, 
                     ten_yr_diff, 
                     on ='Temp_Name', 
                     how ='left')
world.info()
```

```python
world.Difference = world.Difference.astype(float)
world.info()
```

```python
#plotting the temperature differences geospatially
fig4, ax4 = plt.subplots(1, 1, figsize = (15,15))
ax5 = fig4.add_axes([.15, .35,.01, .2])

world.plot(column = 'Difference', 
           ax=ax4,
           cax=ax5,
           cmap='autumn_r',
           legend = True,
          missing_kwds= dict(color = "lightgrey",),
          legend_kwds={"shrink":.4, 
                       'label': "Temperature Increase* (Â°C)"})

ax4.patch.set_facecolor('blue')
ax4.patch.set_alpha(.05)
ax4.set_title("Country Temperature Changes (Jan. 1905 -  May 2011)") #adding title
ax4.set_xlabel("Longitude") #adding x axis label
ax4.set_ylabel("Latitude") #adding y axis label
ax4.xaxis.set_major_formatter('{x:1.0f} Â°')
ax4.yaxis.set_major_formatter('{x:1.0f} Â°')
ax4.margins(x=0.01, y=0.01) 

ax4.axhline(y = 0, xmin = .1, color='grey', linestyle = ':') #line indicating the starting point of our data selection
ax4.text(-100, 1, 'Equator')

ax4.axhline(y = 40, color='blue', linestyle = ':') #line indicating the starting point of our data selection
ax4.text(-175, 41, 'Latitude to explore', color='blue')

ax4.text(-145, -60, '*Temp. increases are based \non 10-year average\ntemperature anomalies', size = 8)
pass
# #providing annotation for each country that we do not have temperature data for
# world[~world.Temp_Present].apply(lambda x: ax3.annotate(text=x['name'],
#                                                    xy=x.geometry.centroid.coords[0],
#                                                    xytext = (x.geometry.centroid.coords[0][0]+5,x.geometry.centroid.coords[0][1]+12),
#                                                    arrowprops = {'facecolor': 'black', 'width':.01},
#                                                    ha='center'), axis=1);
```
